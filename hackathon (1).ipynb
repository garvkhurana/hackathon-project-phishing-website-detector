```python
x=df.drop(columns='result')
y=df["result"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=50,test_size=0.4)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
lr=LogisticRegression()
lr.fit(x_train,y_train)
prediction=lr.predict(x_test)
print(accuracy_score(y_test,prediction))
print(confusion_matrix(y_test,prediction))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
prediction1=knn.predict(x_test)
print(accuracy_score(y_test,prediction1))
print(confusion_matrix(y_test,prediction1))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(max_depth=10)
dt.fit(x_train,y_train)
prediction2=dt.predict(x_test)
print(accuracy_score(y_test,prediction2))
print(confusion_matrix(y_test,prediction2))
```

```python
from sklearn.ensemble import RandomForestClassifier
rc=RandomForestClassifier(max_depth=12)
rc.fit(x_train,y_train)
prediction3=rc.predict(x_test)
print(accuracy_score(y_test,prediction3))
print(confusion_matrix(y_test,prediction3))
```

```python
from sklearn.svm import SVC
s=SVC()
s.fit(x_train,y_train)
prediction4=s.predict(x_test)
print(accuracy_score(y_test,prediction4))
print(confusion_matrix(y_test,prediction4))
```

```python
from sklearn.linear_model import Perceptron
p=Perceptron()
p.fit(x_train,y_train)
prediction8=p.predict(x_test)
print(accuracy_score(y_test,prediction8))
print(confusion_matrix(y_test,prediction8))
```
